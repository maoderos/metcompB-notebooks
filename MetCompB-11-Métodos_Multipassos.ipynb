# Métodos Multipassos 

## Método explicito ou de Adam-Bashforth

$$ x_{n+1} = x_n + \Delta t \left(\frac{3}{2}f(t_n,x_n) - \frac{1}{2}f(t_{n-1},x_{n-1})\right) $$

### Aplicação - Oscilador de Van de Pool

$$ \frac{d^2x}{dt^2}=\mu(1-x^2)\frac{dx}{dt}-x $$

Dividindo em um sistema de equações:

$$ \frac{dx}{dt} = v $$
$$ \frac{dv}{dt} = \mu(1-x^2)v-x$$

```python
import numpy as np
import matplotlib.pyplot as plt

def fX(v):
    return v
def fV(x,v,mu):
    return (mu*(1 - x**2)*v - x)
    
## condições iniciais


tf = 30
dt = 0.001

mu = [0.0,0.1,0.5,1.0,2.0]

for i in mu:
    x0 = 0.5
    v0 = 0
    x1 = x0
    v1 = v0
    t0 = 0
    
    x_l = [x0]
    v_l = [v0]
    t_l = [t0]
    
    while(t0 < tf):
        x2 = x1 + dt*( (3/2)*fX(v1) - (1/2)*fX(v0) )
        v2 = v1 + dt*( (3/2)*fV(x1,v1,i) - (1/2)*fV(x0,v0,i) )
        x_l.append(x2)
        v_l.append(v2)
        x0 = x1
        x1 = x2
        v0 = v1
        v1 = v2
        t0 += dt
        t_l.append(t0)
        
   
    plt.title("x vs v")
    plt.plot(x_l,v_l, label="$\mu = ${0}".format(i))
    plt.legend()
   
   



```
