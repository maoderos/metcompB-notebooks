# Diagrama de Bifurcações

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,a):
    return a*x*(1 - x)

a_list = np.arange(0,4,0.001)

for a in a_list:
    x = 0.2
    a_list = []
    x_list = []
    
    for i in range(200):
        x = f(x,a)
    for i in range(100):
        x = f(x,a)
        x_list.append(x)
        a_list.append(a)
    plt.scatter(a_list,x_list,marker='.',s=1)
    plt.xlim(3,4)
    
```
