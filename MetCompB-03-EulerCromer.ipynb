```python
import numpy as np 
import matplotlib.pyplot as plt

def f(x,w):
  return -(w**2)*x


x_o = 2 
v_o = 0
w = 1
dt = 0.1

t = np.arange(0,30,dt)

x_eulerExplicit = [x_o]
x_eulerCrammer = [x_o]
v_eulerExplicit = [v_o]
v_eulerCrammer = [v_o]

E_crammer = []
E_explicit = [] 
for i in t:
  if i == 0:
    #E = (1/2)*(v_o**2) + (1/2)*
    pass
  else:
    #Met. Euler exp
  
    x_aux = x_eulerExplicit[-1]
    
    v_aux = v_eulerExplicit[-1]

    x_explicit = x_aux + v_aux*dt
  
    v_explicit = v_aux + f(x_aux,w)*dt

    x_eulerExplicit.append(x_explicit)
    v_eulerExplicit.append(v_explicit)
    
    #Met. Euler Crammer
    x_crammer = x_eulerCrammer[-1] +  v_eulerCrammer[-1]*dt
    v_crammer = v_eulerCrammer[-1] + f(x_crammer,w)*dt
   
    x_eulerCrammer.append(x_crammer)
    v_eulerCrammer.append(v_crammer)
    

plt.figure()
plt.plot(t, v_eulerExplicit ,label="v(t) - Explicito")
plt.plot( t, v_eulerCrammer ,label="v(t) - Crammer ")
plt.legend()
plt.show()
plt.close()


plt.figure()
plt.plot(t, x_eulerExplicit ,label="x(t) - Explicito")
plt.plot( t, x_eulerCrammer ,label="x(t) - Crammer ")
plt.legend()
plt.show()
plt.close()
```
