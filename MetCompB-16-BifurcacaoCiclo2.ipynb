# Bifurcação para o ciclo 2

$$g(x) = f(f(x))$$

onde $f(x) = ax(1-x)$

Logo,

$$g(x) = a^2x(1-x)[1 - ax(1-x)]$$

```python
import numpy as np
import matplotlib.pyplot as plt

def id(x):
    return x

def g(x,a):
    return a**2*x*(1 - x)*( 1 - a*x*(1-x) )


x = np.arange(0,1,0.000001)

a = 2.9

plt.plot(x,id(x))
plt.plot(x,g(x,a))

i = 0
while(i < len(x)):
    if (abs(x[i] - g(x[i],a)) < 0.0000001):
        print("intersecção: x = {0}".format(x[i])) 
    i +=1
```
