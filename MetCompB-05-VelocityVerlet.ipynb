# Método de Velocity-Verlet



```python
import numpy as np 
import matplotlib.pyplot as plt

# condições inicias

w=1
x_o=3
v_o=0
dt = 0.005
tf = 50
g = 10
m = 1
l = 10


t = 0
# Metodo de Velocity verlet
t_list = [0]
x_Vverlet = [x_o]
v_Vverlet = [v_o]
E_Vverlet = [(1/2)*m*l**2*v_o**2 + m*g*l*(1 - np.cos(x_o))]
x = x_o 
v = v_o

E_analitico = [(1/2)*m*l**2*v_o**2 + m*g*l*(1 - np.cos(x_o))]

while(t < tf):
  a = -(w**2)*np.sin(x)
  x = x + v*dt - (1/2)*(w**2)*np.sin(x)*dt**2
  a1 = -(w**2)*np.sin(x)
  v = v + (1/2)*(a1 + a)*dt
  En = (1/2)*m*l**2*v**2 + m*g*l*(1 - np.cos(x))
  E_analitico.append((1/2)*m*l**2*v_o**2 + m*g*l*(1 - np.cos(x_o)))
  E_Vverlet.append(En)
  x_Vverlet.append(x)
  v_Vverlet.append(v)
  t += dt
  t_list.append(t)
```

```python
import numpy as np 
import matplotlib.pyplot as plt

# condições inicias

w=1
x_o=3
v_o=0
dt = 0.005
tf = 50
g = 10
m = 1
l = 10

# Primeiro, encontrar x1 utilizando o método de Euler Explicito
x1 = x_o + v_o*dt
t = 0
# Metodo de verlet
t_verlet = [0]
x_arr = [x_o]
v_arr = [v_o]
E_verlet = [(1/2)*m*l**2*v_o**2 + m*g*l*(1 - np.cos(x_o))]
while(t < tf):
  x2 = 2*x1 - x_o - (w**2)*np.sin(x1)*dt**2
  v1 = (x2 - x_o)/(2*dt)
  En = (1/2)*m*l**2*v1**2 + m*g*l*(1 - np.cos(x1))
  E_verlet.append(En)
  x_arr.append(x1)
  v_arr.append(v1)
  x_o = x1
  x1 = x2
  t += dt
  t_verlet.append(t)

```

```python
plt.figure()
plt.title("Energia")
plt.xlabel("t")
plt.ylabel("E")
plt.plot(t_list,E_Vverlet, label="Velocity Verlet")
plt.plot(t_list,E_analitico)
plt.plot(t_verlet,E_verlet, label="Verlet")
plt.legend()
plt.show()
plt.close()

plt.figure()
plt.title("Retrato de fase")
plt.xlabel("t")
plt.ylabel("E")
plt.plot(x_Vverlet,v_Vverlet, label="Velocity Verlet")
plt.plot(x_arr,v_arr, label="Verlet")
plt.legend()
plt.show()
plt.close()
```
