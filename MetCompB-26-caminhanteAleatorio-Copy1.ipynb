```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

class walker():
    x=0.0
    def __init__(self,ident,L):
        self.passo,self.ident=L,ident
    def mov(self):
        self.x+=self.passo*(2*rand.randint(0,1)-1)

L,N=1,100
cam=walker(1,L) #Criando caminhante 1 com passo L
x_list=[0.] #Lista das posicoes
for i in range(N): #Fazendo cam caminhar N passos
    cam.mov()
    x_list.append(cam.x)
plt.plot(x_list)
plt.show()
```

## M√©dia e Desvio Padr√£o

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

class walker():
    x=0.0
    def __init__(self,ident,L):
        self.passo,self.ident=L,ident
    def mov(self):
        self.x+=self.passo*(2*rand.randint(0,1)-1)

L,N=1,100
M = 1000
cam=list(walker(i,L) for i in range(M))#lista de camin.

x_medio = []
x_std = []
n_passos = []
for i in range(N):#Fazendo cam caminhar N passos
    x_m = 0 
    x_standard = 0
    for m in range(M):
        cam[m].mov()
        x_m += cam[m].x/M
    x_medio.append(x_m)
    for m in range(M):
        x_standard += (cam[m].x - x_m)**2/M
    x_std.append(x_standard)
    n_passos.append(i)
    
plt.figure()
plt.plot(n_passos,x_medio,label="caminahante {0}".format(m))
plt.xlabel("Numero de passos")
plt.ylabel("media")
plt.show()
plt.close()
plt.figure()
plt.plot(n_passos,x_std,label="caminahante {0}".format(m))
plt.xlabel("Numero de passos")
plt.ylabel("std")
plt.show()
plt.close()
```
