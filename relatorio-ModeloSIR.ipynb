# Modelo-SIR

$$   \frac{dS}{dt} = -\frac{\beta}{N} IS $$
$$    \frac{dI}{dt} = \frac{\beta I S}{N} $$
$$   \frac{dR}{dt} = \gamma I $$

```python
import numpy as np 
import matplotlib.pyplot as plt

#condicções iniciais
S_o = 99
I_o = 1
R_o = 0
gamma = 0.2
b = 1
N = S_o + I_o + R_o

S_l = [S_o]
I_l = [I_o]
R_l = [R_o]
t_l = [0]
N_l = [N]

S = S_o
I = I_o
R = R_o

tf=20
dt = 0.001
t = dt
#euler-chomer
while t < tf:
  S = S - ((b*I*S)/N)*dt
  I = I + ((b*I*S)/N)*dt - gamma*I*dt
  R = R + gamma*I*dt
  
  Ni = S + I + R

  S_l.append(S)
  I_l.append(I)
  R_l.append(R)
  t_l.append(t)
  N_l.append(Ni)
  t += dt

```

```python
plt.figure()
plt.plot(t_l,S_l, label="S(t)")
plt.plot(t_l,I_l, label="I(t)")
plt.plot(t_l,R_l, label="R(t)")
plt.legend()
plt.show()
plt.close()

plt.figure()
plt.plot(t_l,N_l)
plt.show()
plt.close()
```
