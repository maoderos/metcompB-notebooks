# Modelo-SIR

$$   \frac{dS}{dt} = -\frac{\beta}{N} IS $$
$$    \frac{dI}{dt} = \frac{\beta I S}{N} -\gamma I$$
$$   \frac{dR}{dt} = \gamma I $$

```python
import numpy as np 
import matplotlib.pyplot as plt

#condicções iniciais
S_o = 99
I_o = 1
R_o = 0
gamma = 0.5
b = 1
N = S_o + I_o + R_o

S_l = [S_o]
I_l = [I_o]
R_l = [R_o]
t_l = [0]
N_l = [N]

S = S_o
I = I_o
R = R_o

tf=30
dt = 0.001
t = dt
#euler-chomer
while t < tf:
    S = S - ((b*I*S)/N)*dt
    I = I + ((b*I*S)/N)*dt - gamma*I*dt
    R = R + gamma*I*dt
    Ni = S + I + R
    
    S_l.append(S)
    I_l.append(I)
    R_l.append(R)
    t_l.append(t)
    N_l.append(Ni)
    t += dt

plt.figure()
plt.title(r"$S_o = {0}$, $I_o = {1}$, $R_o = {2}$, $\beta = {3}$, $\gamma = {4}$".format(S_o,I_o,R_o,b,gamma))
plt.ylabel("Indivíduos")
plt.xlabel("t(s)")
plt.plot(t_l,S_l, label="S(t)")
plt.plot(t_l,I_l, label="I(t)")
plt.plot(t_l,R_l, label="R(t)")
plt.plot(t_l, N_l)
plt.legend()
plt.show()
plt.close()

print(N_l[0] - N_l[-1])

plt.title("População \n " + r"$S_o = 99$, $I_o = 1$, $R_o = 0$, $\beta = 1$, $\gamma = 0.5$")
plt.ylabel("Indivíduos")
plt.xlabel("t(s)")
plt.plot(t_l,N_l)
plt.show()
```

## Limites de $S_o$ e $I_o$ para propagação da doença

### Haver propagação 

Para que haja propação:

$$ \frac{\beta S_o}{N} >  \gamma $$

Sabemos que 

$$N = S_o + I_o + R_o$$

mas $R_o = 0$, entao:

$$\beta S_o > \gamma(S_o + I_o)$$

$$(\beta - \gamma)S_o > \gamma I_o$$

Como $\beta = 1.0$ e $\gamma = 0.5$

$$S_o > I_o$$



```python
import numpy as np 
import matplotlib.pyplot as plt

#condicções iniciais
S_o = 85
I_o = 15
R_o = 0
gamma = 0.5
b = 1
N = S_o + I_o + R_o

S_l = [S_o]
I_l = [I_o]
R_l = [R_o]
t_l = [0]
N_l = [N]

S = S_o
I = I_o
R = R_o

tf=30
dt = 0.001
t = dt
#euler-chomer
while t < tf:
    S = S - ((b*I*S)/N)*dt
    I = I + ((b*I*S)/N)*dt - gamma*I*dt
    R = R + gamma*I*dt
    Ni = S + I + R
    
    S_l.append(S)
    I_l.append(I)
    R_l.append(R)
    t_l.append(t)
    N_l.append(Ni)
    t += dt

plt.figure()
plt.title(r"$S_o = {0}$, $I_o = {1}$, $R_o = {2}$, $\beta = {3}$, $\gamma = {4}$".format(S_o,I_o,R_o,b,gamma))
plt.ylabel("Indivíduos")
plt.xlabel("t(s)")
plt.plot(t_l,S_l, label="S(t)")
plt.plot(t_l,I_l, label="I(t)")
plt.plot(t_l,R_l, label="R(t)")
plt.plot(t_l, N_l)
plt.legend()
#plt.show()
plt.savefig("infec.pdf")
plt.close()
```
