```python
import numpy as np 
import matplotlib.pyplot as plt

def f(x,w):
  return -(w**2)*np.sin(x)

def analytic(x_o, t):
  return x_o*np.cos(w*t)

t_o = 0
w = 1
g = 10
m=1
t_f = 24
l=10
dt = 0.1
t_arr = np.arange(0,t_f,dt)

#a =Et_analytic(0.5,t_arr)
#print(a)
x_o = [0.1,0.5,1.0,3.0,3*3.14159, -3*3.14159]
v_o = [0,0,0,0,-0.1,0.1]


for i in range(0,len(x_o)):
  x = x_o[i]
  v = v_o[i]
  t = dt + t_o
  x_eulerCramer = [x_o[i]]
  v_eulerCramer = [v_o[i]]
  E = [(1/2)*m*l**2*v**2 + m*g*l*(1 - np.cos(x))]
  E_an = [(1/2)*m*l**2*v_o[i]**2 + m*g*l*(1 - np.cos(x_o[i]))]
  t_ar = [t_o]
  while(t < t_f):
    x = x + v*dt
    v = v + f(x,w)*dt
    En = (1/2)*m*l**2*v**2 + m*g*l*(1 - np.cos(x))
    x_eulerCramer.append(x)
    v_eulerCramer.append(v)
    E.append(En)
    E_an.append(E_an[0])
    t_ar.append(t)

    t += dt

  
  plt.figure()
  plt.xlabel("t")
  plt.ylabel("Energia (J)")
  plt.title("Energia em função do tempo: x(0) = {0:.2f} e v(0) = {1}".format(x_o[i], v_o[i]))
  plt.plot(t_arr,E_an,label="Analytic")
  plt.plot(t_ar,E,label="Euler-Cramer")
  plt.legend()
  #plt.savefig("E_t_xo_{0}_vo_{1}.pdf".format(x_o[i], v_o[i]))
  plt.show()
  plt.close()



  
```
