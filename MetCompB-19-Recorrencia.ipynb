```python
import matplotlib.pyplot as plt
import numpy as np

def condition(y,z,ro=0.0005):
    if (abs(y - z) < ro):
        return True
    else:
        return False
    


a = 1.2
N = 30
x = 0.2

x_l = [x]
y = []
z = []
for i in np.arange(N):
    x = x*a*(1-x)
    x_l.append(x)
    
for i in np.arange(N):
    for j in np.arange(N):
        if condition(x_l[i],x_l[j]):
            y.append(i)
            z.append(j)

plt.scatter(y,z,s=3)
    
```
