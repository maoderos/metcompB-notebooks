# Expoente de Lyiapunov

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,a):
    return a*x*(1 - x)

def f_derivada(x,a):
    return  a*(1-2*x

a_list = np.arange(0.0001,4,0.001)
a_l = []
lyapunov = []
n = 1000

for a in a_list:
    x = 0.2
    ly = 0
    for i in range(n):
        x = f(x,a)
        x_derivada = f_derivada(x,a)
        ly += np.log(abs(x_derivada))
    lyapunov.append(ly/n)
    a_l.append(a)

plt.scatter(a_l,lyapunov,marker='o',s=1)
```
