# Método de Euler Explícito

Dada a equação: 

$$\frac{dx}{dt} =  f(x,t)$$

Com $x(t=0) = x_o$. Encontrar $x(t)$ que satisfaça a equação para qualquer t.

Pela expansão em série de Taylor:

$$ x(t + \Delta t) = x(t) + \frac{dx}{dt}\Delta t + \frac{1}{2}\frac{d^2x}{dt^2}\Delta t^2 + ...$$

Dado que $\frac{dx}{dt} = f(x,t)$, temos para $t=0$:

$$ x(\Delta t)  = x(0) + f(x,0)\Delta t$$
$$  x(2\Delta t) = x(\Delta t) + f(x,\Delta t) \Delta t$$

Generalizando 

$$ x((i + 1)\Delta t) = x_{i+1} = x_i + f(x_i,i\Delta t)\Delta t$$

```python
import numpy as np 
import matplotlib.pyplot as plt

def analytical_model(x_o,t,tau):
  return x_o*np.exp(-t/tau)

def euler_explicit(x,t,tau,dt):

  return x - (x/tau)*dt 

tau = 2
x_o = 100
tf = 10

```

```python

dt = 0.1
t = np.arange(0,10,dt) 
x_analytic = analytical_model(x_o,t,tau)
x_euler = []

for t_i in t: 
  if t_i == 0:
    x_euler.append(x_o)
  else:
    x_i = euler_explicit(x_euler[-1],t_i,tau,dt)
    x_euler.append(x_i)
    
  

plt.plot(t,x_analytic,label="analytical model")
plt.plot(t,x_euler,label="euler explicit model")
plt.legend()
```
