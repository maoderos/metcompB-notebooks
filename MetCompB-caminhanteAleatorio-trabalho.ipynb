## M√©dia e Desvio Padr√£o

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

class walker():
    x=0.0
    def __init__(self,ident,L):
        self.passo,self.ident=L,ident
    def mov(self):
        self.x+=(rand.uniform(-L,L))
        
def std_analitico(N):
    return N*8.33
L,N=5,1000
M = 10000
cam=list(walker(i,L) for i in range(M))#lista de camin.

x_medio = []
x_std = []
n_passos = []
for i in range(N):#Fazendo cam caminhar N passos
    x_m = 0 
    x_standard = 0
    for m in range(M):
        cam[m].mov()
        x_m += cam[m].x/M
    x_medio.append(x_m)
    for m in range(M):
        x_standard += ((cam[m].x)**2)/M
    x_std.append(x_standard)
    n_passos.append(i)
    
    
std_anali = list(std_analitico(i) for i in n_passos)
plt.figure()
plt.plot(n_passos,x_medio,label="caminahante {0}".format(m))
plt.title("M = {0}".format(M))
plt.xlabel("N√∫mero de passos")
plt.ylabel("M√©dia")
plt.savefig("mean_more.pdf")
plt.close()
plt.figure()
plt.plot(n_passos,x_std,label="Dado".format(m))
plt.plot(n_passos, std_anali,label="Analitico")
plt.legend()
plt.title("M = {0}".format(M))
plt.xlabel("N√∫mero de passos")
plt.ylabel("Desvio quadr√°tico m√©dio")
plt.savefig("std_more.pdf")
plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rand

class walker():
    x=0.0
    def __init__(self,ident,L):
        self.passo,self.ident=L,ident
    def mov(self):
        self.x+=self.passo*(2*rand.randint(0,1)-1)

L,N=1,20
M = 10000
cam=list(walker(i,L) for i in range(M))#lista de camin.
for i in range(N):
    for j in range(M):
        cam[j].mov()
y=[]
for i in cam:
    y.append(i.x)

hist=plt.hist(y,bins=10)
plt.show()
```
